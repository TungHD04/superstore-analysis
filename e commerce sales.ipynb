# Data Storytelling Projects with Superstore

## Step 1: Load & Explore Data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding='latin-1') 
```

## Step 2: Data Cleaning

### Converting Date Columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date']) 

```

### Adding New Date-Based Columns

```python
data['Order Month'] = data['Order Date'].dt.month 
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek

```

### Shipping Time

```python
# Create Delivery Time Column (Number of Days Between Order Date and Ship Date)
data['Shipping Time (days)'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
data.head()
```

## Step 3: Thematic Analysis

### Region and State Sales Analysis

```python
# Total revenue and profit by region
region_summary = data.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

# Create Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=region_summary, x='Region', y='Sales', color='skyblue', label='Sales')
sns.barplot(data=region_summary, x='Region', y='Profit', color='orange', label='Profit')
plt.title('Revenue and profit by Region')
plt.ylabel('Value ($)')
plt.legend()
plt.tight_layout()
plt.show()

# Top 10 states according to profits
state_summary = data.groupby('State')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False).head(10)

# Create Bar Chart for Top 10 States
plt.figure(figsize=(12, 6))
sns.barplot(data=state_summary.reset_index(), x='Profit', y='State', palette='viridis')
plt.title('Top 10 most profitable states')
plt.xlabel('Profit ($)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### Monthly Sales Analysis

```python
# Total Sales by Month (Regardless of Year)
monthly_sales = data.groupby('Order Month')['Sales'].sum().reset_index()

# Sales by Month Chart
plt.figure(figsize=(10, 5))
sns.barplot(x='Order Month', y='Sales', data=monthly_sales, palette='Blues_d')
plt.title(' Total Sales by Month (Aggregated Across All Years)')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.show()
```

```python
# Pivot Table to Create Sales by Month & Year
heatmap_data = data.pivot_table(index='Order Year', columns='Order Month', values='Sales', aggfunc='sum')

# Create Heatmap Chart
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title(' Sales Heatmap by Year and Month')
plt.xlabel('Month')
```

### Sales Analysis by Category

```python
category_sales = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
# Group total sales by category
category_sales = data.groupby('Category')['Sales'].sum().reset_index()

# Set colors
colors = ['#72C1C8', '#F4CF6F', '#F89C7E']  

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Donut chart
wedges, texts, autotexts = ax.pie(
    category_sales['Sales'],
    labels=category_sales['Category'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    pctdistance=0.75,
    wedgeprops=dict(width=0.4),
    textprops=dict(color='black', fontsize=12)
)


ax.axis('equal')


plt.title("Sales Analysis by Category", fontsize=16, loc='left')

# Add legend to the right top corner
ax.legend(
    wedges,
    category_sales['Category'],
    title="Category",
    loc='center left',
    bbox_to_anchor=(1, 0.5)
)

# Adjust layout
plt.tight_layout()
plt.show()
```

### Sales Analysis by Sub-Category

```python
# Total Sales and Profit by Sub-Category
subcategory_summary = data.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).reset_index()

# Sales by Sub-Category Chart
plt.figure(figsize=(12, 6))
sns.barplot(data=subcategory_summary, x='Sub-Category', y='Sales', palette='pastel')
plt.xticks(rotation=45)
plt.title(' Sales by Sub-Category')
plt.tight_layout()
plt.show()

```

### Monthly Profit Analysis

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])  # Ensure Dates Are in Correct Format
data['Order YearMonth'] = data['Order Date'].dt.to_period('M')  # Combined Year-Month Format
```

```python
# Group Profit by Year-Month
monthly_profit = data.groupby('Order YearMonth')['Profit'].sum().reset_index()
monthly_profit['Order YearMonth'] = monthly_profit['Order YearMonth'].astype(str)  # Convert Period to str
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_profit, x='Order YearMonth', y='Profit', marker='o', color='teal')
plt.xticks(rotation=45)
plt.title(' Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Profit Analysis by Category

```python
# Group total profit by category
category_sales = data.groupby('Category')['Profit'].sum().reset_index()

# Set colors
colors = ['#72C1C8', '#F4CF6F', '#F89C7E']  

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create donut chart
wedges, texts, autotexts = ax.pie(
    category_sales['Profit'],  
    labels=category_sales['Category'],  
    autopct='%1.1f%%', 
    startangle=90, 
    colors=colors,  
    pctdistance=0.75, 
    wedgeprops=dict(width=0.4),  
    textprops=dict(color='black', fontsize=12) 
)

# Set axis to equal for proper circular shape
ax.axis('equal')

# Move title to the left corner
plt.title("Profit Analysis by Category", fontsize=16, loc='left')  

# Add legend to the right top corner
ax.legend(
    wedges,
    category_sales['Category'],  
    title="Category",  
    loc='upper right', 
    bbox_to_anchor=(1, 1)  
)

plt.tight_layout()
plt.show()
```

### Profit Analysis by Sub-Category

```python
# Profit by Sub-Category Chart
plt.figure(figsize=(12, 6))
sns.barplot(data=subcategory_summary, x='Sub-Category', y='Profit', palette='coolwarm')
plt.xticks(rotation=45)
plt.title(' Profit by Sub-Category')
plt.tight_layout()
plt.show()
```

### Sales and Profit Analysis by Customer Segment

```python
from matplotlib.ticker import FuncFormatter 

```

```python

segment_data = data.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(segment_data))  # Using numpy array for indexing

# Sales bar
plt.bar(index, segment_data['Sales'], width=bar_width, label='Sales', color='#5DADE2')

# Profit bar
plt.bar(index + bar_width, segment_data['Profit'], width=bar_width, label='Profit', color='#48C9B0')

# Format Y-axis to show in millions (e.g., 1M, 2M)
def millions(x, pos):
    return f'{x / 1e6:.1f}M'  # Adjusted for decimal points

# Apply formatter to Y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))  # FuncFormatter is now correctly imported

# Labels & formatting
plt.xlabel('Customer Segment', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.title('Sales and Profit Analysis by Customer Segment', fontsize=14, weight='bold')
plt.xticks(index + bar_width / 2, segment_data['Segment'], fontsize=10)  # Enhanced tick labels
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

### Shipping time vs profit

```python
# Average Profit by Delivery Time
shipping_impact = data.groupby('Shipping Time (days)')['Profit'].mean().reset_index()

# Create Chart
plt.figure(figsize=(10, 5))
sns.lineplot(data=shipping_impact, x='Shipping Time (days)', y='Profit', marker='o')
plt.title(' Average Profit by Delivery Time')
plt.xlabel('Delivery Days')
plt.ylabel('Average Profit ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### analyse sales-to-profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```

```python

# Prepare Dashboard
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(30, 15))  # Increase Overall Size
fig.subplots_adjust(wspace=0.1, hspace=0.4)  # Adjust Spacing Between Bars
fig.suptitle(" Superstore Sales Dashboard", fontsize=20, fontweight='bold')
import matplotlib.gridspec as gridspec
# Use Gridspec to Adjust the Layout Ratio
fig = plt.figure(figsize=(30, 15))
gs = gridspec.GridSpec(2, 3, width_ratios=[1, 2, 1])  # Chart 2 will occupy twice the width

# Chart 1: Sales và Profit theo Region (Plot in Axes[0, 0])
region_summary = data.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
region_summary.set_index('Region')[['Sales', 'Profit']].plot(
    kind='bar',
    ax=axes[0,0],
    color=['skyblue', 'orange']
)

axes[0,0].set_title("Sales and Profit by Region")
axes[0,0].set_ylabel("Amount ($)")
axes[0,0].legend(title='Metrics')

# Chart 2: Monthly Sales Heatmap

# Pivot Data for Heatmap
heatmap_data = data.pivot_table(index='Order Year', columns='Order Month', values='Sales', aggfunc='sum')
# Plot Heatmap in Subplot Axes[0, 1]
sns.heatmap(
    heatmap_data, 
    annot=True, 
    cmap="YlGnBu", 
    ax=axes[0, 1],
    annot_kws={"size": 10, "color": "black"},  # Increase font size and set color to black for better contrast
    cbar_kws={'label': 'Sales ($)'},  # Add label to color bar
    fmt=",.0f"  # Add thousands separator for readability (consolidated here)
)

axes[0, 1].set_title('Heatmap: Sales by Year and Month')
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Year')


#Chart 3:
from matplotlib.patches import Patch
# Group total sales by category
category_sales = data.groupby('Category')['Sales'].sum().reset_index()

# Color for Each Section
colors = ['#72C1C8', '#F4CF6F', '#F89C7E']

# Axis in Subplot Layout: For example, place in Chart 3's axes position[0, 2]
ax = axes[0, 2]

# Create Donut Chart
wedges, texts, autotexts = ax.pie(
    category_sales['Sales'],
    labels=category_sales['Category'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    pctdistance=0.75,
    wedgeprops=dict(width=0.4),
    textprops=dict(color='black', fontsize=8)
)

# Keep Circular Chart
ax.axis('equal')

# Chart Title
ax.set_title("Sales by Category", fontsize=10, loc='left')




# Chart 4: Sales by Sub-Category (in axes[1,0])
subcategory_summary = data.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).reset_index()

sns.barplot(data=subcategory_summary, x='Sub-Category', y='Sales', palette='pastel', ax=axes[1, 0])
axes[1, 0].set_title('Sales by Sub-Category')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].set_ylabel('Sales ($)')
axes[1, 0].set_xlabel('')



# Chart 5: Sales & Profit by Segment (in axes[1,1])
segment_data = data.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

bar_width = 0.35
index = np.arange(len(segment_data))

# Plot on Subplot
axes[1, 1].bar(index, segment_data['Sales'], width=bar_width, label='Sales', color='#5DADE2')
axes[1, 1].bar(index + bar_width, segment_data['Profit'], width=bar_width, label='Profit', color='#48C9B0')

# Format y-axis to millions
def millions(x, pos):
    return f'{x / 1e6:.1f}M'

axes[1, 1].yaxis.set_major_formatter(FuncFormatter(millions))

# Set Labels & Format
axes[1, 1].set_title('Sales and Profit by Customer Segment', fontsize=11)
axes[1, 1].set_xlabel('Segment')
axes[1, 1].set_ylabel('Amount ($)')
axes[1, 1].set_xticks(index + bar_width / 2)
axes[1, 1].set_xticklabels(segment_data['Segment'], fontsize=9)
axes[1, 1].legend(loc='upper right')
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Chart 6: Average Profit by Delivery Time (in axes[1, 2])
shipping_impact = data.groupby('Shipping Time (days)')['Profit'].mean().reset_index()

sns.lineplot(data=shipping_impact, x='Shipping Time (days)', y='Profit', marker='o', ax=axes[1, 2])
axes[1, 2].set_title('Average Profit by Shipping Time')
axes[1, 2].set_xlabel('Shipping Time (days)')
axes[1, 2].set_ylabel('Average Profit ($)')
axes[1, 2].grid(True)
```
